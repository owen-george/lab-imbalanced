# LAB | Imbalanced

**Load the data**

In this challenge, we will be working with Credit Card Fraud dataset.

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv

Metadata

- **distance_from_home:** the distance from home where the transaction happened.
- **distance_from_last_transaction:** the distance from last transaction happened.
- **ratio_to_median_purchase_price:** Ratio of purchased price transaction to median purchase price.
- **repeat_retailer:** Is the transaction happened from same retailer.
- **used_chip:** Is the transaction through chip (credit card).
- **used_pin_number:** Is the transaction happened by using PIN number.
- **online_order:** Is the transaction an online order.
- **fraud:** Is the transaction fraudulent. **0=legit** -  **1=fraud**


```python
#Libraries
import pandas as pd
import numpy as np
import time
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
from sklearn.metrics import fbeta_score, make_scorer, accuracy_score,  confusion_matrix, cohen_kappa_score, precision_score, recall_score, classification_report, f1_score
from sklearn.preprocessing import OneHotEncoder,  StandardScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

from sklearn.utils import resample
from imblearn.over_sampling import SMOTE
```

```python
fraud = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv")
fraud.head()
```

**Steps:**

- **1.** What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?
- **2.** Train a LogisticRegression.
- **3.** Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.
- **4.** Run **Oversample** in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model? 
- **5.** Now, run **Undersample** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?

1. What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?

```python
fraud['fraud'].value_counts()
```

```python
#This dataset is very imbalanced - over 10x fraud==0 vs fraud==1
```

2. Train a LogisticRegression.

```python
target = fraud['fraud']
features = fraud.drop('fraud', axis=1)

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)

#Normalise all columns to be 0-1
normalizer = MinMaxScaler()
normalizer.fit(x_train)

x_train_norm = normalizer.transform(x_train)
x_test_norm = normalizer.transform(x_test)

x_train_norm = pd.DataFrame(x_train_norm, columns=x_train.columns, index=x_train.index )
x_test_norm = pd.DataFrame(x_test_norm, columns=x_test.columns, index=x_test.index)
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train_norm, y_train)
```

3. Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.

```python
pred = log_reg.predict(x_test_norm)
```

```python
print(f"MAE: {mean_absolute_error(y_test, pred): .4f}")
print(f"MSE: {mean_squared_error(y_test, pred): .4f}")
print(f"RMSE: {root_mean_squared_error(y_test, pred): .4f}")
print(f"Recall: {recall_score(y_test, pred): .4f}")
print(f"Precision: {precision_score(y_test, pred): .4f}")
print(f"Kappa: {cohen_kappa_score(y_test, pred): .4f}")
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Pred: No", "Pred: Yes"], yticklabels=["True: No", "True: Yes"])
plt.title("Confusion Matrix for Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print(f"When predicted fraud, this is correct {100* (7112/(7112+628)): .2f}% of the time")
print(f"{100* (7112/(7112+10273)): .2f}% fraud events are predicted")
```

```python
print(f"When predicted fraud, this is correct {100* (cm[1][1]/(cm[1][1]+cm[0][1])): .2f}% of the time")
print(f"{100* (cm[1][1]/(cm[1][1]+cm[1][0])): .2f}% fraud events are predicted")
```

From the above, it is clear most fraud events aren't correctly predicted with the model - skewed by sample imbalance making the "accuracy" result look strong

4. Run Oversample in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model?

```python
train = pd.DataFrame(x_train_norm, columns = x_train.columns)
train["fraud"] = y_train.values

fraud_yes = train[train["fraud"] == 1]
fraud_no = train[train["fraud"] == 0]
```

```python
fraud_oversampled = resample(fraud_yes, 
                    replace=True, 
                    n_samples = len(fraud_no),
                    random_state=0)
```

```python
train_over = pd.concat([fraud_oversampled, fraud_no])
train_over['fraud'].value_counts()
```

```python
x_train_over = train_over.drop(columns = ["fraud"])
y_train_over = train_over["fraud"]


log_reg.fit(x_train_over, y_train_over)
```

```python
pred2 = log_reg.predict(x_test_norm)
```

```python
print(f"MAE: {mean_absolute_error(y_test, pred2): .4f}")
print(f"MSE: {mean_squared_error(y_test, pred2): .4f}")
print(f"RMSE: {root_mean_squared_error(y_test, pred2): .4f}")
print(f"Recall: {recall_score(y_test, pred2): .4f}")
print(f"Precision: {precision_score(y_test, pred2): .4f}")
print(f"Kappa: {cohen_kappa_score(y_test, pred2): .4f}")
```

```python
# Significant increase in Recall score and Kappa
```

```python
cm2 = confusion_matrix(y_test, pred2)

plt.figure(figsize=(6, 5))
sns.heatmap(cm2, annot=True, fmt="d", cmap="Blues", xticklabels=["Pred: No", "Pred: Yes"], yticklabels=["True: No", "True: Yes"])
plt.title("Confusion Matrix for Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print(f"When predicted fraud, this is correct {100* (cm2[1][1]/(cm2[1][1]+cm2[0][1])): .2f}% of the time")
print(f"{100* (cm2[1][1]/(cm2[1][1]+cm2[1][0])): .2f}% fraud events are predicted")
```

```python
# Even thouhgh there are now more false predictions, far fewer fraud events are missed. Therefore this is a big improvement
```

```python
x_train_under = train_under.drop(columns = ["fraud"])
y_train_under = train_under["fraud"]

log_reg.fit(x_train_under, y_train_under)
```

```python
pred3 = log_reg.predict(x_test_norm)
```

```python
print(f"MAE: {mean_absolute_error(y_test, pred3): .4f}")
print(f"MSE: {mean_squared_error(y_test, pred3): .4f}")
print(f"RMSE: {root_mean_squared_error(y_test, pred3): .4f}")
print(f"Recall: {recall_score(y_test, pred3): .4f}")
print(f"Precision: {precision_score(y_test, pred3): .4f}")
print(f"Kappa: {cohen_kappa_score(y_test, pred3): .4f}")
```

```python
cm3 = confusion_matrix(y_test, pred3)

plt.figure(figsize=(6, 5))
sns.heatmap(cm3, annot=True, fmt="d", cmap="Blues", xticklabels=["Pred: No", "Pred: Yes"], yticklabels=["True: No", "True: Yes"])
plt.title("Confusion Matrix for Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print(f"When predicted fraud, this is correct {100* (cm3[1][1]/(cm3[1][1]+cm3[0][1])): .2f}% of the time")
print(f"{100* (cm3[1][1]/(cm3[1][1]+cm3[1][0])): .2f}% fraud events are predicted")
```
